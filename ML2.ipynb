```python
import pandas as pd

df = pd.read_csv(r"C:\Users\padma\Downloads\ML_Lab-2.csv")
display(df)
print("The first five row's:\n", df.head(5))
print("\nThe last 5 rows:\n", df.tail(5))
print("\nThe No.of rows and columns:",df.shape)
print("The names of columns:\n",df.columns)
print("The Datatypes of the columns:\n",df.dtypes)
df.info()
print("Students with attendance above 90:\n", df[df['Attendance'] > 90])
```

```python
import numpy as np
print("The Mean of the dataset is", np.mean(df['Attendance']))
print("\nThe average of the internalmarks:", np.average(df['InternalMarks']))
print("The Maximum of the Internalmarks:",np.max(df['InternalMarks']))
print("The Minimum of the Internalmarks:", np.min(df['InternalMarks']))
print("\nThe average if the externalmarks:", np.average(df['ExternalMarks']))
print("The Maximum of the Externalmarks:" ,np.max(df['ExternalMarks']))
print("The Minimum of Externalmarks:", np.min(df['ExternalMarks']))
print("\nThe Standard deviation of the Internal Marks:", np.std(df['InternalMarks']))
df.describe()
```

```python
import numpy as np
df['TotalMarks']  = df['InternalMarks'] + df['ExternalMarks']
df['Percentage'] = (df['TotalMarks']/100)*100
display(df)

print("Students with attendance above 90:\n", df.loc[df['Attendance'] > 90])
print("Students whose total marks exceeds 90:\n",df.loc[df['TotalMarks']>90])
print("Female Students:\n",df.loc[df['Gender']=='F'])
print("Male Students:\n",df.loc[df['Gender']=='M'])
print("Students with above average marks:\n",df[df['TotalMarks']>np.average(df['TotalMarks'])])
```

```python
print("Male Students:\n",df.loc[df['Gender']=='M'])
print("Students with above average marks:\n",df[df['TotalMarks']>np.average(df['TotalMarks'])])
print("Average Marks :", np.average(df["TotalMarks"]))
```

```python
import numpy as np

avg_male = np.average(df.loc[df['Gender'] == 'M', 'TotalMarks'])
print("The Averange Marks of the Male Students", avg_male)
avg_female = np.average(df.loc[df['Gender'] == 'F', 'TotalMarks'])
print("The Averange Marks of the Female Students", avg_female)
avg_attend_m = np.average(df.loc[df['Gender']=='M', 'Attendance'])
print("The average attendance of Male", avg_attend_m)
avg_attend_f = np.average(df.loc[df['Gender']=='F', 'Attendance'])
print("The average attendance of Female", avg_attend_f)
print("The greater attendance in male and female is", max(avg_attend_f,avg_attend_m))
```

```python
import matplotlib.pyplot as plt
plt.bar(df['Name'], df['TotalMarks'])
plt.xlabel('Students')
plt.ylabel('Total Marks')
plt.show()

plt.hist(df['Attendance'],bins=10)
plt.xlabel('Attendance Range (%)')
plt.ylabel('Number of Students')
plt.show()

gender_type = df['Gender'].value_counts()
plt.pie(gender_type, labels=gender_type.index)
plt.show()

plt.scatter(df['Attendance'],df['TotalMarks'])
plt.xlabel('Attendance')
plt.ylabel('Total Marks')
plt.show()

print("The Relation Between Attendance and Total Marks is Positive")
print("\nThe Female students have better attendance and better total marks than the male students.")
print("\nThe Female Students performed well in exams than the male students according to their averages in total marks.")
print("\nMaximum no.of students have marks above the average of total marks (i.e 85.1).")
print("\nThe Higher Attendance is increasing with the better academic performance")
```
